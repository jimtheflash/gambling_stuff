```python
import nba_api.stats.endpoints.scoreboardv2 as sb2
import pandas as pd
from datetime import timedelta, date
import time
import os

start_date = date.today()
end_date = date(2021, 8, 1)
delta = timedelta(days=1)

schedule_list = []

while start_date <= end_date:
    print(start_date)
    time.sleep(2)
    date_string = str(start_date)
    scoreboard = sb2.ScoreboardV2(game_date = date_string)
    scoreboard = sb2.ScoreboardV2()
    all_dfs = scoreboard.get_data_frames()
    schedule_df = all_dfs[0]
    schedule_list.append(schedule_df)
    start_date += delta

schedule_df = pd.concat(schedule_list)
```
